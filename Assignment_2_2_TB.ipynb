**Part 2:**
In this assignment, you will complete a notebook where you will build a classifier to predict whether a loan case will be paid off or not.  

Assumption: Principal is quoted in thousands so 1000 is actually 1M

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset1 = pd.read_csv('loan_train.csv')
dataset1.head()
```

```python
for col in ['due_date', 'effective_date']:
    dataset1[col] = pd.to_datetime(dataset1[col]).apply(lambda date: date.toordinal())
```

```python
X1 = dataset1.iloc[:, 3:].values
y1 = dataset1.iloc[:, 2].values
print(y1[0:5])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [5,6])], remainder='passthrough')
X1 = np.array(ct.fit_transform(X1))
print(X1[0:3, :])
```

```python
for i in range(len(y1)):
    if y1[i] == 'PAIDOFF':
        y1[i] = 1
    else:
        y1[i] = 0

y1 = y1.astype(int)
print(np.unique(y1))
print(y1[1])
```

**Splitting the data**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.2, random_state = 0)
```

**Scaling the data**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Classification models**  
**K_Nearest Neighbour**

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

**Testing Classification Model: KNN**
Remember to convert date

```python
#Change values for eff_date, due_date
from datetime import datetime # Import the datetime module

eff_date ="09/10/1985"
eff_date = datetime.strptime(eff_date, "%m/%d/%Y").toordinal() # Convert the string to a datetime object first, then get the ordinal

due_date ="09/20/1985"
due_date = datetime.strptime(due_date, "%m/%d/%Y").toordinal() # Convert the string to a datetime object first, then get the ordinal


print(eff_date,",",due_date)
```

```python
#enter format: education, gender, principal, terms, effective date, due date, age. Use eff_date and due_date values printed above.
#Education format (first four entries): bachelor, college, high school, masters. Add 1 in the respective column, and 0 for the rest.
#Gender format (next 2 entries): female, male. Add 1 in the respective column, 0 for the other.
classifier.predict(sc.transform([[0,1,0,0,1,0,1000,30,724894, 724904, 39, ]]))
```

Based on the reading, a college educated female with a $1M ($1000 K) principal and a 30 year term, purchasing the house with an effective date of 09/10/1985 and due date of 09/20/1985 **would pay off** their house.

**Checking the accuracy score of the model**

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

Not bad for a model. True diagonal (TP, TN) > false diagonal (FP, FN), with an accuracy score of 0.77.
